## Analysis

### Some Observations made from the analysis include:
### There were more male than female mice in the study 

### The mice were tested most on Capomulin

### The Drug Regimen, Infubinol had the highest total tumor volume


```python
# Import Dependencies 
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
```

```python
# files to load 
file_1="Resources/Mouse_metadata.csv"
file_2="Resources/Study_results.csv"

# read in files 
mouse_data=pd.read_csv(file_1)
study_results=pd.read_csv(file_2)

# Combine the data 
data=pd.merge(study_results,mouse_data, how="left", on=["Mouse ID", "Mouse ID"])
data.head()
```

```python
# Checking the number of mice.
mice_count=len(data["Mouse ID"].unique())
mice_count
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse UD and Timepoint 
timepoint=data["Timepoint"].value_counts()
time_id_appears=data["Mouse ID"].value_counts()
#reset the index
time_id_appears=time_id_appears.reset_index()
duplicate_mice_row=time_id_appears.loc[time_id_appears['Mouse ID']>len(timepoint)]
duplicate_mice_id=duplicate_mice_row.iloc[0,0]
duplicate_mice_id
```

```python
# Optional: Get all the data for the duplicate mouse ID.
data_for_duplicate_mice=data.loc[data['Mouse ID']==duplicate_mice_id]
data_for_duplicate_mice
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_data=data.loc[data["Mouse ID"]!=duplicate_mice_id]
clean_data
```

```python
# Checking the number of mice in the clean DataFrame.
new_mice_count=len(clean_data['Mouse ID'].unique())
new_mice_count
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
mean_tv=clean_data.groupby(['Drug Regimen']).mean()['Tumor Volume (mm3)']
median_tv=clean_data.groupby(['Drug Regimen']).median()['Tumor Volume (mm3)']
variance_tv=clean_data.groupby(['Drug Regimen']).var()['Tumor Volume (mm3)']
std_tv=clean_data.groupby(['Drug Regimen']).std()['Tumor Volume (mm3)']
sem_tv=clean_data.groupby(['Drug Regimen']).sem()['Tumor Volume (mm3)']
# Assemble the resulting series into a single summary DataFrame.
summary_df=pd.DataFrame({'Mean Tumor Volume':mean_tv,
                        'Median Tumor Volume':median_tv,
                        'Tumor Volume Variance':variance_tv,
                        'Tumor Volume Std. Dev.':std_tv,
                        'SEM':sem_tv})
summary_df
```

## Quartiles, Outliers and Boxplots

```python
#Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
greatest_timepoint=clean_data.groupby(['Mouse ID'])
greatest_timepoints=greatest_timepoint['Timepoint'].max()
greatest_timepoints=greatest_timepoints.reset_index()
greatest_timepoint_df=pd.DataFrame(greatest_timepoints)
greatest_timepoint_df

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merged_df=greatest_timepoint_df.merge(clean_data, on=['Mouse ID','Timepoint'], how='left')
merged_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments=['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
t_volume=[]

# Calculate the IQR and quantitatively determine if there are any potential outliers. 

    
# Locate the rows which contain mice on each drug and get the tumor volumes
for drug in treatments:
    tumor_by_drug=merged_df.loc[merged_df['Drug Regimen']==drug,'Tumor Volume (mm3)']
    t_volume.append(tumor_by_drug)
#print(t_volume)
    
# add subset 
    quartiles=merged_df['Tumor Volume (mm3)'].quantile([.25,.5,.75])
    lowerq=quartiles[0.25]
    upperq=quartiles[0.75]
    iqr=upperq-lowerq
# Determine outliers using upper and lower bounds
    lowerbound=lowerq-(1.5*iqr)
    upperbound=upperq+(1.5*iqr)
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
fig1, ax1=plt.subplots()
x=[1,2,3,4]
ax1.set_title('Tumor Volume by Drug Regimen')
ax1.set_ylabel('Tumor Volume (mm3)')
ax1.set_xlabel('Drug Regimen')
ax1.boxplot(t_volume)
plt.xticks(x,treatments)
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
capomulin_mouse=clean_data.loc[clean_data['Mouse ID']=='l509']
#capomulin_mouse
capomulin_time=capomulin_mouse['Timepoint']
capomulin_volume=capomulin_mouse['Tumor Volume (mm3)']
x_axis=np.arange(0,45,0.5)
plt.plot(capomulin_time, capomulin_volume)
plt.xlabel('Timepoint')
plt.ylabel('Tumor Volume (mm3)')
plt.title('Capomulin Treatment of Mouse l509')
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
capomulin_drug=clean_data.loc[clean_data['Drug Regimen']=='Capomulin']
volume_mean=capomulin_drug.groupby(['Mouse ID']).mean()
tumor_vol_mean=volume_mean['Tumor Volume (mm3)']
weight_mean=volume_mean['Weight (g)']
#plot scatter plot
plt.scatter(weight_mean, tumor_vol_mean, marker='o', facecolors='blue', edgecolor='black')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
correlation=st.pearsonr(weight_mean, tumor_vol_mean)
print(f'The correlation between mouse weight and average tumor volume is {round(correlation[0],2)}')

(slope, intercept,rvalue,pvalue,stderr)=linregress(weight_mean, tumor_vol_mean)
regress=slope*weight_mean+intercept
line_eq=f'y={round(slope,2)} x {round(intercept,2)}'

plt.scatter(weight_mean,tumor_vol_mean)
plt.plot(weight_mean,regress,'r-')
plt.annotate(line_eq,(20,36),fontsize=15,color='red')


plt.xlabel('Weight (g)')
plt.ylabel('Tumor Volume (mm3)')
plt.title('Regression by Average Mouse Weight and Average Tumor Volume')
plt.show()
```
